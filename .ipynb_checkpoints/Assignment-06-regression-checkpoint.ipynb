The linear model on top of a scatter plot for your training and test data

```python
model = sklearn.linear_model.LinearRegression()

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    x[:,2], y, test_size=0.2, random_state=42)

x_train_trans = x_train.reshape(-1,1)
x_test_trans = x_test.reshape(-1,1)

model.fit(x_train_trans,y_train)


x_model_vals = np.linspace(-0.1, 0.2, 50).reshape(-1,1)
y_model_vals = model.predict(x_model_vals)
plt.plot(x_model_vals ,y_model_vals,'green')

plt.scatter(x_train_trans, y_train, color='blue')
plt.scatter(x_test_trans, y_test, color='red')

y_pred = model.predict(x_test_trans)
```

The coefficient of the model

```python
print(model.coef_)
```

The mean squared error and R-squared values for your model applied to the test data

```python
from sklearn.metrics import mean_squared_error, r2_score
print('MSE_linreg = %.2f' % mean_squared_error(y_test, y_pred))
print('R^2 = %.2f' % r2_score(y_test, y_pred))
```

```python
model2 = sklearn.linear_model.LinearRegression()

x_train2, x_test2, y_train2, y_test2 = sklearn.model_selection.train_test_split(
    x, y, test_size=0.2, random_state=42)


model2.fit(x_train2 ,y_train2)


y_pred2 = model2.predict(x_test2)

print('MSE_linreg = %.2f' % mean_squared_error(y_test2, y_pred2))
print('R^2 = %.2f' % r2_score(y_test2, y_pred2))
```

The mean squared error for my model applied to the test data calculated with all 10 features is significantly lower than that with only one feature of the data, and the former model's $ R^2 $ is smaller than the latter model's.  The smaller MSE is preferred because it indicates that my data points are dispersed closely around the predicted values, and that my estimation is more accurate. And the fact that $ R^2 $ of the latter model is closer to 1 implies the latter model better predicts the distribution of the data.

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
        x[:,2],y, test_size=0.2, random_state=42)

x_train_trans = x_train.reshape(-1,1)


k_range = range(1, 60)
k_scores = []
for k in k_range:
    knn = sklearn.neighbors.KNeighborsRegressor(n_neighbors=k)
    loss = cross_val_score(knn,
                           x_train_trans,
                           y_train, 
                           cv=5, 
                           scoring='neg_mean_squared_error')
    k_scores.append(np.sqrt(-loss).mean())
plt.scatter(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated MSE')
plt.show()
```

```python
k_range = range(40, 50)
k_scores = []
for k in k_range:
    knn = sklearn.neighbors.KNeighborsRegressor(n_neighbors=k)
    loss = cross_val_score(knn,
                           x_train_trans,
                           y_train, 
                           cv=5, 
                           scoring='neg_mean_squared_error')
    k_scores.append(np.sqrt(-loss).mean())
plt.scatter(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated MSE')
plt.show()
```
